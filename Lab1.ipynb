# 3. Regular Expressions
A regular expression is a piece of code (written in the standardized regular expression language) that specifies a textual pattern. For instance, I could use natural language to refer to "the set of all words beginning with the letter a," however (as EECS 487 discusses extensively) this is very difficult for a computer to parse. Regular expressions enable us to formally specify textual patterns.

Regular expressions enable programs to quickly search large bodies of text for particular patterns. They can enable tasks such as information extraction (finding and returning important pieces of information to the user/program), word replacement, and many others.

One interesting application of regular expressions is [ELIZA](https://web.njit.edu/~ronkowit/eliza.html), an early chatbot (created in 1966!) that imitated a Rogerian psychotherapist. Although relatively simple under the hood, ELIZA is a convincing dialogue partner.

Python's [standard library](https://docs.python.org/3/library/index.html) comes preinstalled with `re`, a package for defining and using regular expressions. Here is [a link](https://docs.python.org/3/library/re.html) to the package documentation. Throughout this tutorial we will learn how to define regular expressions using `re`, and we will conclude with an exercise giving you the opportunity to apply your knowledge.

While working with `re`, it may be helpful to use a [regex cheatsheet](https://images.datacamp.com/image/upload/v1665049611/Marketing/Blog/Regular_Expressions_Cheat_Sheet.pdf). This cheatsheet provides an overview of the types of patterns you can specify using `re`.

````python
import re
````

````python
# the search function allows us to find the first location of the pattern
pattern = 'oppy'
test_str = 'my dog is named poppy and she is very hoppy'
x = re.search(pattern, test_str)
print(x)

# findall allows us to find all instances of the pattern
x = re.findall(pattern, test_str)
print(x)

# sub allows us to substitute all instances of the pattern
new_str = 'ip'
x = re.sub(pattern, new_str, test_str)
print(x)
````

````python
# the ^ character stands for the start
# the . character stands for any character
pattern = '^.'
x = re.findall(pattern, test_str)
print('Results of using ^')
print(x)

# the $ character stands for the end
pattern = '.$' # If you use $., you get nothing-- there is no character after the end of the string
x = re.findall(pattern, test_str)
print('Results of using $')
print(x)
````

````python
test_str = 'I have 2 pet rabbits'

# \d can be used to stand for any number
pattern = '\d'
x = re.findall(pattern, test_str)
print(x)

# \D can be used to stand for non-numeric characters
pattern = '\D\D'
x = re.findall(pattern, test_str)
print(x)

# \w can be used to stand for any character inside of a word
pattern = '\w\w'
x = re.findall(pattern, test_str)
print(x)
````

````python
test_str = 'There are approximately 2^5 different species of rabbits'

# use backslashes to find special characters
pattern = '\d\^\d'
x = re.search(pattern, test_str)
print(x)
````

````python
test_str = 'This is my rabbit Harrison'

# + can be used to match strings with characters that repeat >=1 times
pattern = 'r+.'
x = re.findall(pattern, test_str)
print(x)

# * can be used to match strings with characters that repeat >=0 times
pattern = 'ab*'
x = re.findall(pattern, test_str)
print(x)
````

````python
test_str = 'This is a rabbit named Harrison'

# use (?:x) syntax to find patterns but not include them
pattern = '(?:a)b*'
x = re.findall(pattern, test_str)
print(x)

# use (x|y) syntax to find patterns of one or the other type
pattern = '(a)(bb|rr)'
x = re.findall(pattern, test_str)
print(x)

# use (?<!x) syntax to exclude certain characters x
# \s refers to whitespace
pattern = '(?<!\s)a..'
x = re.findall(pattern, test_str)
print(x)
````

## Exercise 3

For this activity, you will play the part of a secret agent responsible for declassifying a large set of documents. You have to remove all of the dates, times, and email addresses from these documents. There is no need to remove agent names since your agency uses codenames anyways. To speed up the process, you decide to use regular expressions to automatically remove these types of strings from your data.

Please create a string `scrubbed_text` that contains the "cleaned" contents of `secret_agent_task.txt`. All names, dates, times, and email addresses should be replaced with the string `REDACTED`.

[RegExr](https://regexr.com/) is nice visual tool to workshop regex patterns that you may wish to try!

````python
with open('secret_agent_task.txt') as fp:
    secret_text = fp.read()

# feel free to use these two strings for your function:
months = '(January|February|March|April|May|June|July|August|September|October|November|December)'
replace_str = 'REDACTED'

def scrub_text(text):
    # Remove dates
    date_pattern_1 = months + "\s\d*(\,\s\d{4})*"
    text = re.sub(date_pattern_1, replace_str, text)
    date_pattern_2 = "\d{2}\/\d{2}(\/\d{4})*"
    text = re.sub(date_pattern_2, replace_str, text)

    # Remove times
    time_pattern = "\d{2}\:\d{2}"
    text = re.sub(time_pattern, replace_str, text)

    # Remove emails
    email_pattern = "\S*\@\S*"
    text = re.sub(email_pattern, replace_str, text)

    return text

print('ORIGINAL SECRET TEXT:')
print(secret_text)
scrubbed_text = scrub_text(secret_text)
print('\nSCRUBBED TEXT:')
print(scrubbed_text)
````
